```python
df=pd.read_csv('employees.csv')
df.head()
```

```python
"""
Question 6:
    a) Shape (number of rows and columns) of the DataFrame.
    b) Summary of the DataFrame that includes the data types and non-null counts for each column.
    c) Generate descriptive statistics.
    d) Display the first 5 rows and last 3 rows of the dataset.
    e) Calculate the following statistics from the dataset:
        i. The average salary of employees.
        ii. The total bonus paid to all employees.
        iii. The youngest employee's age.
        iv. The highest performance rating.
    f) Sort the DataFrame by the Salary column in descending order.
    g) Add a new column that categorizes employees based on their performance rating:
        i. Excellent for ratings >= 4.5
        ii. Good for ratings >= 4.0 but < 4.5
        iii. Average for ratings < 4.0
    h) Identify missing values in the DataFrame.
    i) Rename the Employee_ID column to ID.
    j) Find all employees who:
        i. Have more than 5 years of experience.
        ii. Belong to the IT department.
    k) Modify the dataset by adding a new column, Tax, which deducts 10% of the Salary.
    l) Save the modified DataFrame (with added columns) to a new CSV file.
"""
```

```python
# a) Shape (number of rows and columns) of the DataFrame.

print("Shape of the Dataframe is: ",df.shape)
```

```python
# b) Summary of the DataFrame that includes the data types and non-null counts for each column.

print("Summary of the DataFrame is:")
df.info()
```

```python
# c) Generate descriptive statistics.

print("Descriptive Statistics of the Dataframe is:")
df.describe()
```

```python
# d) Display the first 5 rows and last 3 rows of the dataset.

print("First five rows:")
df.head(5)
```

```python
# d) Display the first 5 rows and last 3 rows of the dataset.

print("\nLast three rows are:")
df.tail(3)
```

```python
"""
e) Calculate the following statistics from the dataset:
        i. The average salary of employees.
        ii. The total bonus paid to all employees.
        iii. The youngest employee's age.
        iv. The highest performance rating.
"""

print("Average Salary of Employee is: ",df['Salary'].mean())
print("Total Bonus paid to all Employee is: ",df['Bonus'].sum())
print("Youngest Employee's age is: ",df['Age'].min())
print("Highest performance rating: ",df['Rating'].max())
```

```python
# f) Sort the DataFrame by the Salary column in descending order.

print("\nDataframe sorted by Salary in descending order:")
df.sort_values(by='Salary',ascending=False)
```

```python
"""
g) Add a new column that categorizes employees based on their performance rating:
        i. Excellent for ratings >= 4.5
        ii. Good for ratings >= 4.0 but < 4.5
        iii. Average for ratings < 4.0
"""

def categorize_rating(rating):
    if rating >= 4.5:
        return "Excellent"
    elif 4.0 <= rating < 4.5:
        return "Good"
    else:
        return "Average"

df["Performance_Category"] = df["Rating"].apply(categorize_rating)
print("DataFrame with Performance Category:")
df.head()
```

```python
# h) Identify missing values in the DataFrame.

print("Missing values in the DataFrame:")
print(df.isnull().sum())
```

```python
# i) Rename the Employee_ID column to ID.

df.rename(columns={'Employee_ID':'ID'})
print("DataFrame after rename Employee_ID column to ID:")
df.head()
```

```python
"""
j) Find all employees who:
        i. Have more than 5 years of experience.
        ii. Belong to the IT department.
"""

print("Employees who have more than 5 years of experience:")
df[df['Years_of_Experience']>5]
```

```python
"""
j) Find all employees who:
        i. Have more than 5 years of experience
        ii. Belong to the IT department.
"""

print("Employees who belong to IT department:")
df[df['Department']=='IT']
```

```python
# k) Modify the dataset by adding a new column, Tax, which deducts 10% of the Salary. 

df['Tax']=df['Salary']*0.1
print("DataFrame after adding new column Tax:")
df.head()
```

```python
# l) Save the modified DataFrame (with added columns) to a new CSV file. 

df.to_csv("employees_modified.csv", index=False)
print("\nModified DataFrame saved to 'employees_modified.csv'.")
```
