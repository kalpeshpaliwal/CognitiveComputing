```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""
Question 1:
For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'), Find
    i. Sum of all elements 
    ii. Sum of all elements row-wise 
    iii. Sum of all elements column-wise 
"""
gfg = np.matrix('[4,1,9; 12,3,1; 4,5,6]')
print("\nGFG Matrix:\n",gfg)

print("\nSum of all elements: ",np.sum(gfg))
print("\nSum of all elements row-wise:\n",np.sum(gfg,axis=1))
print("\nSum of all elements column-wise:\n",np.sum(gfg,axis=0))
```

```python
"""
Question 2:
(a)For the array: array = np.array([10, 52, 62, 16, 16, 54, 453]), find 
        i. Sorted array 
        ii. Indices of sorted array 
        iii. 4 smallest elements 
        iv. 5 largest elements 
"""
array = np.array([10, 52, 62, 16, 16, 54, 453])
print("Array                        : ", array)

sort_array = np.sort(array)
indices_sorted = np.argsort(array)
print("Sorted array                 : ", sort_array)
print("Indices of sorted array      : ", indices_sorted)
print("Four smallest elements       : ", sort_array[0:4])         #np.partition(array, 4)[:4]
print("Five largest elements        : ", sort_array[-1:-6:-1])    #np.partition(array, -5)[-5:]
```

```python
"""
Question 2:
(b) For the array: array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0]), find 
        i. Integer elements only 
        ii. Float elements only 
"""
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
print("Array                        : ", array)

int_element = array[array == array.astype(int)]
float_element = array[array != array.astype(int)]
print("Integer Elements Only        : ", int_element)
print("Float Elements Only          : ", float_element)
```

```python
"""
Question 3:
You are given a weekly sales dataset and need to perform various operations
using NumPy broadcasting.

a) Generate your unique sales dataset:

        Take the sum of the ASCII values of the initials of your first and last
        name. Call this value X. (If your initials are A B → ASCII 
        sum = 65 + 66 = 131 → sales = [131, 181, 231, 281, 331].)
        
        Create a NumPy array sales with values [X, X+50, X+100, X+150, X+200].

b) Compute your personalized tax rate as ((X % 5) + 5) / 100.
        Use broadcasting to apply this tax rate to each sales value.
        
c) Adjust sales based on discount:
        If sales < X+100, apply a 5% discount.
        If sales >= X+100, apply a 10% discount. 
        
d) Expand sales data for multiple weeks:
        Create a 3×5 matrix representing three weeks of sales by stacking sales three times using broadcasting.
        Increase sales by 2% per week using element-wise broadcasting.
"""
initials = "KS"
X = sum(ord(char) for char in initials)
sales = np.array([X, X+50, X+100, X+150, X+200])
print("Sales                            : ", sales)

tax_rate = ((X % 5) + 5) / 100
taxed_sales = sales * (1 + tax_rate)
print("Taxed Sales                      : ", taxed_sales)

discounted_sales = np.where(sales < X + 100, sales * 0.95, sales * 0.90)
print("Discounted Sales                 : ", discounted_sales)

sales_matrix = np.tile(sales, (3, 1))
week_multipliers = np.array([1.00, 1.02, 1.04]).reshape(3, 1)
weekly_sales = sales_matrix * week_multipliers
print("Weekly Sales Data                :\n", weekly_sales)
```

```python
"""
Question 4:
Generate x values using np.linspace() from -10 to 10 with 100 points. Use
each function from the list below and compute y values using NumPy:
    Y = x2
    Y = sin(x)
    Y = ex
    Y = log(|x| + 1)
Plot the chosen function using Matplotlib. Add title, labels, and grid for clarity.
"""

x = np.linspace(-10, 10, 100)

y1 = x**2                       

plt.figure(figsize=(10, 6))
plt.plot(x, y1, label="Y = x²")
plt.title("Plot of Y = x²")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.grid(True)
plt.show()
```

```python
y2 = np.sin(x)

plt.plot(x, y2, label="Y = sin(x)", linestyle="--", color="g")
plt.title("Plot of Y = sin(x)")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.grid(True)
plt.show()
```

```python
y3 = np.exp(x)  

plt.plot(x, y3, label="Y = exp(x)", color="r")
plt.title("Plot of Y = exp(x)  ")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.grid(True)
plt.show()
```

```python
y4 = np.log(np.abs(x) + 1) 

plt.plot(x, y4, label="Y = log(|x| + 1)", color="y")
plt.title("Plot of Y =  log(|x| + 1) ")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.grid(True)
plt.show()
```
