```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
"""
Question 1:
Ask the user to enter a value (e.g., M) for a mathematical function. 
    -> Generate x values from -10 to 10 using np.linspace(). 
    -> Compute y values for: 
            1) y= M⋅x2 
            2) y=M⋅sin(x) 
    -> Plot both functions in a single figure: 
            1) Use different colors and line styles. 
            2) Add legend, grid, and title. 
"""

M = float(input("Enter value for M: "))
x = np.linspace(-10, 10, 50)
y1 = M * x ** 2
y2 = M * np.sin(x)
plt.plot(x, y1, 'r--', label='y=M*x^2')
plt.plot(x, y2, 'g-', label='y=M*sin(x)')
plt.legend()
plt.grid()
plt.title("Mathematical Functions")
plt.show()
```

```python
"""
Question 2:
Create a dataset of five subjects and scores. 
    -> Convert it into a Pandas DataFrame. 
    -> Plot the scores using a Seaborn bar plot with: 
            1) Different colors for each bar. 
            2) Annotations on top of each bar. 
            3) Title, axis labels, and grid. 
"""

subjects = ['Math', 'Science', 'English', 'History', 'Computer Science']
scores = [85, 90, 78, 88, 95]

df = pd.DataFrame({'Subject': subjects, 'Score': scores})
df
```

```python
plt.figure(figsize=(8, 5))
ax = sns.barplot(x='Subject', y='Score', data=df, hue = 'Subject', palette='plasma', legend = False)

for index, row in df.iterrows():
    ax.text(index, row.Score + 1, str(row.Score), color='black', ha="center")

plt.title("Student Scores in Different Subjects")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.ylim(0, 100)  

plt.grid(axis='y', linestyle='--')

plt.show()
```

```python
"""
Question 3:
Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().
        Create a 2x2 subplot layout displaying:
                Line plot showing cumulative sum.
                Scatter plot with random noise.
        Customize with titles, axis labels, and grids.
"""

roll_number = 102314045 
np.random.seed(roll_number)

data = np.random.randn(50)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(np.cumsum(data), linestyle='-', color='b')
axes[0, 0].set_title("Cumulative Sum Line Plot")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].grid(True, linestyle='--')

x_values = np.arange(50)
noise = np.random.randn(50) * 0.5  # Adding some noise
y_values = data + noise
axes[0, 1].scatter(x_values, y_values, color='r')
axes[0, 1].set_title("Scatter Plot with Random Noise")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Value + Noise")
axes[0, 1].grid(True, linestyle='--')

axes[1, 0].plot(data) 
axes[1, 0].set_title('Line Plot')
axes[1, 0].set_xlabel('Index')
axes[1, 0].set_ylabel('Values')
axes[1, 0].grid(True)

axes[1, 1].scatter(np.arange(50), data) 
axes[1, 1].set_title('Scatter Plot')
axes[1, 1].set_xlabel('Index')
axes[1, 1].set_ylabel('Values')
axes[1, 1].grid(True)


plt.tight_layout()
plt.show()
```

```python
"""
Question 4:
Download Data-set from the below link
https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv

Apply ‘seaborn’ library to do the following.
    1. Read Total profit of all months and show it using a line plot.
    2. Read all product sales data and show it using a multiline plot.
    3. Plot bar chart for all the features/attributes.
"""

df = pd.read_csv('company_sales_data.csv')

plt.figure(figsize=(10, 6))
sns.lineplot(x='month_number', y='total_profit', data=df, color='b')
plt.title('Total Profit of All Months')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

```python
product_columns = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

plt.figure(figsize=(12, 8))
for product in product_columns:
    sns.lineplot(x='month_number', y=product, data=df, label=product)

plt.title('Sales data of all Products')
plt.xlabel('Month')
plt.ylabel('Sales Units')
plt.legend(title='Products')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize = (12,8))

for feature in df.columns[1:]:
    plt.figure()
    sns.barplot(x = df.index+1, y = feature, data = df)
    plt.xlabel('Month')
    plt.ylabel(feature)
    plt.grid(axis = 'y)
    plt.show()
```
